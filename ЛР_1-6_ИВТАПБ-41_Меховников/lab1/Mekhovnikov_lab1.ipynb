```python
def age_category(age):
    '''
    < 30 -> 1
    >= 30, < 55 -> 2
    >= 55 -> 3
    '''
    if age < 30:
        return 1
    elif age < 55:
        return 2
    elif age >= 55:
        return 3
```

```python
age_categories = [age_category(age) for age in data.Age]
data['Age_catergory'] = age_categories

#data['Age_catergory'] = data['Age'].apply(age_category)
```

```python
#1. Сколько мужчин / женщин было на борту?
count_male = len(data[data['Sex'] == 'male'])
count_female = len(data[data['Sex'] == 'female'])
print(f"{count_male} мужчин и {count_female} женщин")
```

```python
#2. Определите распределение функции Pclass. 
#Теперь Для мужчин и женщин отдельно. Сколько 
#людей из второго класса было на борту?

count_second_class_male = len(data[(data['Pclass'] == 2) & (data['Sex'] == 'male')])
count_second_class_female = len(data[(data['Pclass'] == 2) & (data['Sex'] == 'female')])

print(f"Мужчин из 2-го класс: {count_second_class_male}\nЖенщин из 2-го класса: {count_second_class_female}")
```

```python
# 3. Каковы медиана и стандартное отклонениеFare?. 
# Округлите до 2-х знаков после запятой.
print(f"медиана {round(data['Fare'].describe().loc['50%'],2)}, "
      f"стандартное отклонение {round(data['Fare'].describe().loc['std'],2)}")
```

```python
#** 4. Правда ли, что средний возраст выживших 
# людей выше, чем у пассажиров, которые в конечном 
# итоге умерли? **

survived = data[data['Survived'] == 1]
no_survived = data[data['Survived'] == 0]

mean_age_survived = survived['Age'].describe().loc['mean']
mean_age_no_survived = no_survived['Age'].describe().loc['mean']

if mean_age_survived > mean_age_no_survived:
    print("Да")
else:
    print("Нет")
```

```python

# 6. Правда ли, что женщины выживали чаще мужчин?
#Каковы доли выживших людей среди мужчин и женщин?

male = len(data[data['Sex'] == 'male'])
female = len(data[data['Sex'] == 'female'])

male_survived = len(data[(data['Sex'] == 'male') &
                     (data['Survived'] == 1)])

female_survived = len(data[(data['Sex'] == 'female') &
                       (data['Survived'] == 1)])

print(f"{round((male_survived/male)*100,1)}% среди мужчин, "
      f"{round((female_survived/female)*100,1)}% среди женщин")
```

```python
#7. Какое имя наиболее популярно среди пассажиров мужского пола?
male = data[data['Sex'] == 'male']
name_male = male['Name']
name = []
middle_name = []

for n in name_male:
    full_name = n.split()
    name.append(full_name[2])
    if len(full_name) > 3:
        middle_name.append(full_name[-1])

name_count = {}

for i in range(len(name)):
    if name[i] in name_count:
        name_count[name[i]] += 1
    else:
        name_count[name[i]] = 1

print(max(name_count, key=name_count.get))
```
